```python
X = df.drop(['Customer Name','Customer e-mail','Country','Car Purchase Amount'],axis=1)
y = df['Car Purchase Amount']
```

```python
X
```

```python
y
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mm = MinMaxScaler()
X_scaled = mm.fit_transform(X)
```

```python
X_scaled
```

```python
y_reshape = y.values.reshape(-1,1)
```

```python
mm1 = MinMaxScaler()
y_scaled = mm1.fit_transform(y_reshape)
```

```python
y_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)

svm = SVR()
svm.fit(X_train, y_train)

rf = RandomForestRegressor()
rf.fit(X_train, y_train)

gb = GradientBoostingRegressor()
gb.fit(X_train, y_train)

xgb = XGBRegressor()
xgb.fit(X_train, y_train)

```

```python
import tensorflow.keras
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

```python
ann = Sequential()
```

```python
ann.add(Dense(30, input_dim=5, activation='relu'))
```

```python
ann.add(Dense(30, activation='relu'))
```

```python
ann.add(Dense(1, activation="linear"))
```

```python
ann.summary()
```

```python
ann.compile(loss='mean_squared_error', optimizer=Adam())
```

```python
ann.fit(X_train, y_train, epochs=100, verbose=1, batch_size=50, validation_split=0.2)
```

```python
y_pred1 = lr.predict(X_test)
y_pred2 = svm.predict(X_test)
y_pred3 = rf.predict(X_test)
y_pred4 = gb.predict(X_test)
y_pred5 = xgb.predict(X_test)
y_pred6 = ann.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
s1 = r2_score(y_test, y_pred1)
s2 = r2_score(y_test, y_pred2)
s3 = r2_score(y_test, y_pred3)
s4 = r2_score(y_test, y_pred4)
s5 = r2_score(y_test, y_pred5)
s6 = r2_score(y_test, y_pred6)
```

```python
print(s1, s2, s3, s4, s5, s6)
```

```python
final_data =pd.DataFrame({'model':['lr','svm','rf','gb','sgb','ann'],
                          'score':[s1, s2, s3, s4, s5, s6]})
```

```python
final_data
```

```python
sns.barplot(data=final_data, x='model',y='score', palette='viridis')
```
