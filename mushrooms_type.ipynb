```python
X = df.iloc[:, 1:23]
y = df.iloc[:,1]
```

```python
X
```

```python
y
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
sc = StandardScaler()
x_scaled = sc.fit_transform(X)
```

```python
x_scaled
```

```python
pca = PCA(n_components=12)
```

```python
X_pca = pca.fit_transform(x_scaled)
```

```python
X_pca
```

```python
X_pca.shape
```

```python
pca.explained_variance_ratio_
```

```python
sum(pca.explained_variance_ratio_)
```

```python

from sklearn.model_selection import KFold , cross_val_score
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
# Set up k-fold cross-validation
kf = KFold(n_splits=10, shuffle=True, random_state=42)

# Define models
models = {
    'RandomForest': RandomForestClassifier(),
    'LogisticRegression': LogisticRegression(max_iter=1000),
    'SVM': SVC(),
    'KNN': KNeighborsClassifier(),
    'GradientBoosting': GradientBoostingClassifier(),
    'DecisionTree': DecisionTreeClassifier()
}

# Define scoring metrics
scoring_metrics = ['accuracy', 'precision_macro', 'recall_macro', 'f1_macro']

# Store results in a DataFrame
results = []

# Evaluate each model with each scoring metric
for model_name, model in models.items():
    for scoring in scoring_metrics:
        scores = cross_val_score(model, X_pca, y, cv=kf, scoring=scoring)
        results.append({
            'Model': model_name,
            'Metric': scoring,
            'Mean Score': np.mean(scores),
            'Std Dev': np.std(scores)
        })

# Convert results to DataFrame for better visualization
results_df = pd.DataFrame(results)

# Print the results
print(results_df)
```

```python
results_df.sort_values(by='Mean Score', ascending=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a bar plot for the scores
plt.figure(figsize=(16, 10))
sns.barplot(y='Mean Score', x='Model', hue='Metric', data=results_df, ci=None)
plt.title('Model Comparison Across Different Metrics')
plt.xlabel('Mean Score')
plt.ylabel('Model')
plt.legend(title='Metric')
plt.show()
```

```python
svm = SVC()
model = svm.fit(X_pca, y)
```

```python
import pickle

with open('models/mushroom.pkl', 'wb') as file:
    pickle.dump(model, file)
    
with open('models/mushroom.pkl', 'rb') as file:
    loaded_model = pickle.load(file)
```
