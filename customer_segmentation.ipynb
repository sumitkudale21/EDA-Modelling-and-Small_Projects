```python
df2 = df.iloc[:,3:5]
```

```python
df2
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans()
km.fit(df2)
```

```python
km.predict(df2)
```

```python
wcss = []
for i in range(1,11):
    km = KMeans(n_clusters=i)
    km.fit(df2) 
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11), wcss)
plt.title("Elbow Method")
plt.xlabel("number of clusters")
plt.ylabel("wcss")
plt.show()
```

```python
k_mean = KMeans(n_clusters=5, random_state=43)
pred = k_mean.fit_predict(df2)
pred
```

```python
plt.scatter(df2.iloc[pred==0,0], df2.iloc[pred==0,1], s=100, color='red', label='cluster_0')
plt.scatter(df2.iloc[pred==1,0], df2.iloc[pred==1,1], s=100, color='yellow', label='cluster_1')
plt.scatter(df2.iloc[pred==2,0], df2.iloc[pred==2,1], s=100, color='green', label='cluster_2')
plt.scatter(df2.iloc[pred==3,0], df2.iloc[pred==3,1], s=100, color='blue', label='cluster_3')
plt.scatter(df2.iloc[pred==4,0], df2.iloc[pred==4,1], s=100, color='purple', label='cluster_4')
plt.scatter(k_mean.cluster_centers_[:,0 ], k_mean.cluster_centers_[:,1], s=100, color='magenta')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
```
