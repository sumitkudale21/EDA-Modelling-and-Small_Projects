```python
text = "Renewable energy sources have gained significant attention in recent years due to their potential to mitigate climate change and reduce dependency on fossil fuels. Solar energy, for instance, harnesses the power of sunlight through photovoltaic cells and thermal collectors, converting it into electricity or heat. Wind energy utilizes turbines to capture the kinetic energy of wind, generating electricity without emissions. Biomass and biofuels derive energy from organic materials like plants and agricultural waste, offering sustainable alternatives to traditional fuels. Hydroelectric power harnesses the energy of flowing water through dams and turbines, providing a reliable source of electricity. Geothermal energy taps into heat stored beneath the Earth's surface for heating and electricity production. These renewable sources not only diversify energy portfolios but also contribute to global efforts towards a cleaner, more sustainable future. Governments and industries worldwide are increasingly investing in research, development, and infrastructure to expand renewable energy capacity, aiming to reduce carbon footprints and ensure energy security for future generations."
```

```python
len(text)
```

```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from string import punctuation
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
doc = nlp(text)
```

```python
print(doc.text)
for token in doc:
    print(token.text, token.pos_, token.dep_)
```

```python
tokens = [token.text.lower() for token in doc if not token.is_stop and not token.is_punct and token.text != '\n']
```

```python
tokens
```

```python
from collections import Counter
```

```python
word_freq = Counter(tokens)
```

```python
max_freq = max(word_freq.values())
```

```python
for word in word_freq.keys():
    word_freq[word] = (word_freq[word]/max_freq)
```
